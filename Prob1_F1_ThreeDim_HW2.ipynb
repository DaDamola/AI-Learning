```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt


def f1(x, y):
    return 1 - abs(x + y) - abs(y - x)

x = np.linspace(0, 50, 50) 
y = np.linspace(0, 50, 40)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)

p1Z = Z.min()

res = np.where(Z == p1Z)
coords = list(zip(res[0], res[1]))

row = coords[0][0]
col = coords[0][1]

p1X = X[row][col]
p1Y = Y[row][col]

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

ax.plot([p1X], p1Y, p1Z,'ro') 
ax.text(p1X, p1Y, p1Z, "p1", fontsize=12)









```
