```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f3(x, y):
    return x * y 

x = np.linspace(0, 50, 50) 
y = np.linspace(0, 50, 40)

X, Y = np.meshgrid(x, y)
Z = f3(X, Y)

grad = np.gradient(Z) 
res, coords = (), []

for narr in grad:
  if 0 in narr:
   res = np.where(narr == 0)
   coords = list(zip(res[0], res[1]))
   break

row = coords[0][0]
col = coords[0][1]

p3X = X[row][col]
p3Y = Y[row][col]
p3Z = Z[row][col]

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

ax.plot([p3X], p3Y,'ro') 
ax.text(p3X, p3Y, p3Z, "p3", fontsize=12)
```
